```python
import pandas as pd
import numpy as np
data_path = "cleaned_data.csv"
df = pd.read_csv(data_path)
```

```python
def OlderSibling(row):
    "This function determines who is an older sibling"
    if row['OlderBrothers'] == '0' and row['OlderSisters'] == 0:
        return 'Older'
    elif row['OlderBrothers'] != '0' or row['OlderSisters'] != 0:
        return 'Younger'
    else:
        return 'Only Child'

df['OlderSibling'] = df.apply(OlderSibling, axis=1)
df.head()
```
